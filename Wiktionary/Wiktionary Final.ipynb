```python
import mechanize
import http.cookiejar
from bs4 import BeautifulSoup
import re

import pandas as pd
import time
```

```python
# Browser
br = mechanize.Browser()

# Cookie Jar
cj = http.cookiejar.LWPCookieJar()
br.set_cookiejar(cj)

# Browser options
br.set_handle_equiv(True)
br.set_handle_gzip(True)
br.set_handle_redirect(True)
br.set_handle_referer(True)
br.set_handle_robots(False)
br.set_handle_refresh(mechanize._http.HTTPRefreshProcessor(), max_time=1)

br.addheaders = [('User-agent', 'Chrome')]
```

```python
start_time = time.time()

# URL to load
url = 'https://mn.wiktionary.org/w/index.php?title=Ангилал:Монголоор&from=Ё'

# Loaded html
html = br.open(url).read()

# Read html with Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Find and print title to check we have loaded correctly
print(soup.find('h1').string.strip())

# List to contain words
words = []

count = 0
while (soup.find(id='mw-pages').find_all('a')[1].string == 'дараагийн хуудас'):# and (count < 5):
    
    # Add new words to list
    for element in soup.find('div', id='mw-pages').contents[9].contents[0].find_all('a'):
        words.append(element.string.strip())
    
    # Update URL
    url = soup.find(id='mw-pages').find_all('a')[1]['href']
    
    # Loaded html
    html = br.open(url).read()

    # Read html with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    
    count += 1
    print(len(words), 'words,', time.time() - start_time, 'seconds')

words = pd.Series(words).str.lower().drop_duplicates()

time.time() - start_time
```

```python
words.to_excel('wiktionary.xlsx', index=False)
```
